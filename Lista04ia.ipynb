```python
from imdb import IMDb
```

```python
# criando instância da classe IMDb
ia = IMDb()
```

```python
def validarEntradas(artista_nome, ano_inicial, ano_final):
  if (artista_nome == '' or ano_inicial == '' or ano_final == ''):
    print('Entrada vazia')
```

```python
#@title **Filtro de Pesquisa**

#@markdown Preencha os seguintes requisitos para realizarmos a busca:

#@markdown 1. Nome do ator ou atriz:
artista_nome = 'Chris Evans' #@param {type: "string"}

#@markdown 2. Ano de lançamento (anos) 

#@markdown Informe o período que deseja procurar: 

#@markdown Entre
ano_inicial = 1997  #@param {type:"integer"}

#@markdown e
ano_final = 2021 #@param {type:"integer"}

#@markdown 3. Gênero: 
titulo_genero = "Action" #@param ["Action", "Adventure", "Animation", "Biography", "Comedy", "Crime", "Documentary", "Drama", "Family", "Fantasy", "Film Noir", "History", "Horror", "Music", "Musical", "Mystery", "Romance", "Sci-Fi", "Short Film", "Sport", "Superhero", "Thriller", "War", "Western"]

#@markdown 4. Classificação/ pontuação: 
titulo_pontuacao = 7.8 #@param {type:"slider", min:0, max:10, step:0.1}

#@markdown 5. Premiação: 
titulo_premiacao = True #@param {type:"boolean"}

validarEntradas(artista_nome, ano_inicial, ano_final)
```

```python
# pegar id do artista

dados_artista = ia.search_person(artista_nome) 
id_artista = dados_artista[0].personID
```

```python
# Retornar a filmografia do artista
filmografia_artista = ia.get_person(id_artista)
```

```python
def filtrarTitulos(filmografia_artista):

  chave = "actress" if 'actress' in filmografia_artista['filmography'].keys() else "actor"

  for titulo in filmografia_artista['filmography'][chave]:
    if 'year' in titulo.keys() and titulo.data['year'] >= ano_inicial and titulo.data['year'] <= ano_final:
        titulo_dados_completos = ia.get_movie(titulo.getID())
        if titulo_genero in titulo_dados_completos['genres']:
          if 'rating' in titulo_dados_completos.keys() and titulo_dados_completos['rating'] == titulo_pontuacao:
            if titulo_premiacao == True:
              premiacao = ia.get_movie_awards(titulo.getID())
              if (premiacao['data'] != {}):
                return titulo_dados_completos['title']
            else:
              return titulo_dados_completos['title']

```

```python
def arquivar(objeto_indicacao):
  with open('perfil.txt','a') as dados: 
        dados.write(str(objeto_indicacao) + "\n")
```

```python
# Imprimir título indicado e solicitar ao usuário para marcar se curtiu ou não a indicação

indicacao = filtrarTitulos(filmografia_artista)
print("Indicação:", indicacao)

if (indicacao):
  curtiu = int(input("Informe se curtiu a indicação colocando 1:Sim / 0:Não "))
  objeto_indicacao = {
    "titulo_indicado": indicacao,
    "curtiu": str(curtiu),
  }
  arquivar(objeto_indicacao)
```
